```python
df_merged = pd.merge(df_2022, df_2023, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2021, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2020, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2019, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2018, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2017, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2016, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2015, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2014, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2013, on='FIPS', how='left')

df_merged
```

```python
output_directory = 'output_directory_path'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
    
output_filename = 'percent_children_poverty_data.xlsx'
output_file = os.path.join(output_directory, output_filename)

df_merged.dropna(subset=['FIPS'], inplace=True)
df_merged['FIPS'] = df_merged['FIPS'].astype(str).str.zfill(5)
columns = df_merged.columns.tolist()
columns.insert(4, columns.pop(1))
df_merged = df_merged[columns]
start_pos = 3
end_pos = 13
columns = df_merged.columns.tolist()
subset_columns = columns[start_pos:end_pos+1]
subset_columns.reverse()
columns[start_pos:end_pos+1] = subset_columns
df_merged = df_merged[columns]
df_merged.iloc[:, 3:14] = df_merged.iloc[:, 3:14].round(1)
df_merged.to_excel(output_file, index=False)
df_merged.fillna(0, inplace=True)
df_merged = df_merged[df_merged.iloc[:, 2] != 0]
df_merged = df_merged.replace(' ', '_', regex=True)
df_merged = df_merged.replace('% Children in Poverty', '%_Children_in_Poverty', regex=True)

# df_merged.to_sql('df_merged', e, if_exists='append')

df_merged
```

```python
X = df_merged.iloc[:, 3:12]  
y = df_merged['% Children in Poverty_2023']
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r_squared = r2_score(y_test, y_pred)
print('R-squared:', r_squared)
```

```python
output_directory = 'output_directory_path'
output_filename = 'predictions_%_Children_in_Poverty_2023.xlsx'
output_file = os.path.join(output_directory, output_filename)
```

```python
predictions_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
predictions_df.to_excel(output_file, index=False)
print(f"Predictions exported to: {output_file}")
```

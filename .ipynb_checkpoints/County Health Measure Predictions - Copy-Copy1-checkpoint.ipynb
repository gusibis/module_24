```python
year_2023 = '2023'

year_2022 = '2022'

year_2021 = '2021'

year_2020 = '2020'

year_2019 = '2019'

year_2018 = '2018'

year_2017 = '2017'

year_2016 = '2016'

year_2015 = '2015'

year_2014 = '2014'

year_2013 = '2013'
```

```python
column_name= '% Adults with Diabetes'

new_column_name_2023 = f'{year_2023}_{column_name}'  # Modify the column name accordingly
df_2023.rename(columns={column_name: new_column_name_2023}, inplace=True)

new_column_name_2022 = f'{year_2022}_{column_name}'  # Modify the column name accordingly
df_2022.rename(columns={column_name: new_column_name_2022}, inplace=True)

new_column_name_2021 = f'{year_2021}_{column_name}'  # Modify the column name accordingly
df_2021.rename(columns={column_name: new_column_name_2021}, inplace=True)

new_column_name_2020 = f'{year_2020}_{column_name}'  # Modify the column name accordingly
df_2020.rename(columns={column_name: new_column_name_2020}, inplace=True)

new_column_name_2019 = f'{year_2019}_{column_name}'  # Modify the column name accordingly
df_2019.rename(columns={column_name: new_column_name_2019}, inplace=True)

new_column_name_2018 = f'{year_2018}_{column_name}'  # Modify the column name accordingly
df_2018.rename(columns={column_name: new_column_name_2018}, inplace=True)

new_column_name_2017 = f'{year_2017}_{column_name}'  # Modify the column name accordingly
df_2017.rename(columns={column_name: new_column_name_2017}, inplace=True)

new_column_name_2016 = f'{year_2016}_{column_name}'  # Modify the column name accordingly
df_2016.rename(columns={column_name: new_column_name_2016}, inplace=True)

new_column_name_2015 = f'{year_2015}_{column_name}'  # Modify the column name accordingly
df_2015.rename(columns={column_name: new_column_name_2015}, inplace=True)

new_column_name_2014 = f'{year_2014}_{column_name}'  # Modify the column name accordingly
df_2014.rename(columns={column_name: new_column_name_2014}, inplace=True)

new_column_name_2013 = f'{year_2013}_{column_name}'  # Modify the column name accordingly
df_2013.rename(columns={column_name: new_column_name_2013}, inplace=True)
```

```python
df_merged = pd.merge(df_2022, df_2023, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2021, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2020, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2019, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2018, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2017, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2016, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2015, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2014, on='FIPS', how='left')
df_merged = pd.merge(df_merged, df_2013, on='FIPS', how='left')

df_merged
```

```python
df_merged.dropna(subset=['FIPS'], inplace=True)

df_merged['FIPS'] = df_merged['FIPS'].astype(str).str.zfill(5)

df_merged['2013_% Adults with Diabetes'] = df_merged['2013_% Adults with Diabetes'].round(1)

# Get the column names of the DataFrame
columns = df_merged.columns.tolist()

# Swap the positions of the columns
columns.insert(4, columns.pop(1))

# Reorder the DataFrame columns
df_merged = df_merged[columns]

df_merged
```

```python
start_pos = 3
end_pos = 13

columns = df_merged.columns.tolist()

subset_columns = columns[start_pos:end_pos+1]

subset_columns.reverse()

columns[start_pos:end_pos+1] = subset_columns

df_merged = df_merged[columns]

df_merged
```

```python
X = df_merged.iloc[:, 3:13]  
y = df_merged['2023_% Adults with Diabetes']  
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

```python
output_directory = 'output_directory_path'
output_filename = 'predictions.xlsx'
output_file = os.path.join(output_directory, output_filename)
```

```python
predictions_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
predictions_df.to_excel(output_file, index=False)
print(f"Predictions exported to: {output_file}")
```

```python
# Predict values for 2024
X_pred = df_merged.iloc[:, 3:13]  
X_pred_encoded = pd.get_dummies(X_pred, drop_first=True)
y_pred_2024 = model.predict(X_pred_encoded)
```

```python
year_2024 = '2024'
new_column_name_2024 = f'{year_2024}_{column_name}'  
df_merged[new_column_name_2024] = y_pred_2024
```

```python
import pandas as pd
import os
```

```python
directory = 'Recources'
```

```python
excel_files = [file for file in os.listdir(directory) if file.endswith('.xlsx')]
```

```python
combined_data = pd.DataFrame()
```

```python
for file in excel_files:
    # Read the Excel file and get the first sheet
    df = pd.read_excel(os.path.join(directory, file))
    
    # Concatenate the data with the combined data
    combined_data = pd.concat([combined_data, df], ignore_index=True)
```

```python
combined_data
```

```python
output_directory = 'output_directory_path'
```

```python
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
output_filename = 'combined_data.xlsx'
```

```python
output_file = os.path.join(output_directory, output_filename)
```

```python
combined_data.to_excel(output_file, index=False)
```

```python
print(f"Combined data exported to: {output_file}")
```
